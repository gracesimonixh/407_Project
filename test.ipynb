```python
import pandas as pd

from portfolio import Portfolio
```

```python
tickers = ['AAPL', 'SPY']

equity_curve = pd.DataFrame(columns=['Date', 'Cash', 'Portfolio_Value'] + tickers)

equity_curve.head()
```

```python
p = Portfolio(start_cash=10000)
prices_day1 = pd.Series({'AAPL': 150, 'SPY': 400})
p.buy('AAPL', shares=10, price=150, date='2026-02-01') 
p.buy('SPY', shares=5, price=400, date='2026-02-01')  
p.update(prices_day1)
p.record_equity(date='2026-02-01')
print("Cash:", p.cash)
print("\nPositions:")
print(p.positions)

print("\nTrades:")
for t in p.trades:
    print(t)

print("\nEquity Curve:")
print(p.equity_curve)
```
