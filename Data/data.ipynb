```python
import yfinance as yf
import pandas as pd
```

```python
tickers = ["AAPL", "SPY", "JNJ"]

df = yf.download(
    tickers,
    start="2020-01-01",
    end="2023-01-01",
    auto_adjust=True
)

df_clean = (
    df
    .stack(level=1)
    .rename_axis(["Date", "Ticker"])
    .reset_index()
    .sort_values("Date")
    .dropna()
    .reset_index(drop=True)
)

df_clean.columns.name = None

df_clean.to_csv("clean_stock_data.csv", index=False)
```

```python
df_close = df_clean.pivot(index="Date", columns="Ticker", values="Close").reset_index()
df_close = df_close.sort_values("Date")  

df_close.to_csv("date_close_data.csv", index=False)
```
